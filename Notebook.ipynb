## **Notebook: Naturaleza de los materiales dielectricos**

Se denomina dieléctrico al material mal conductor de electricidad,
por lo que puede ser utilizado como aislante eléctrico.


Aunque el dieléctrico y el aislador generalmente se consideran sinónimo, el término dieléctrico es más de uso frecuente al considerar el efecto de alternar campos eléctricos en la sustancia mientras que el aislador es más de uso frecuente cuando el material se está utilizando para soportar un alto campo eléctrico.

#Descripción teórica:

Las moléculas polares bajo la acción de un campo eléctrico experimentan un par de fuerzas que tienden a orientarlas en el sentido del campo. Las moléculas no polares, se hacen polares en presencia de un campo eléctrico, ya que las fuerzas sobre cada tipo de carga son iguales y de sentido contrario.

Un dipolo eléctrico es un sistema formado por dos cargas iguales **q** y de signo contrario, separadas una distancia **d**. Se define el momento dipolar **p**, como un vector cuyo módulo es el producto de la carga **q** por la separación entre cargas **d**, de dirección la recta que las une, y de sentido de la negativa a la positiva.

# Naturaleza (solidos,gases y liquidos):

Los dieléctricos sólidos son quizás los dieléctricos más de uso general de la ingeniería eléctrica, y muchos sólidos son aislantes muy buenos.


Por su naturaleza el aire, nitrógeno y hexafluoruro del sulfuro son los tres dieléctricos gaseosos más de uso general. Los gases aislantes más utilizados en los transformadores son el aire y el nitrógeno, este último a presiones de 1 atmósfera. Estos transformadores son generalmente de construcción sellada. 

Las propiedades físicas de los dieléctricos líquidos como por ejemplo: peso específico, conductibilidad térmica, calor específico, constante dieléctrica, viscosidad, dependen de su naturaleza, es decir de la composición química, pero su rigidez dieléctrica, además está ligada a factores externos como por ejemplo: impureza en suspensión, en solución, humedad, etc., que, generalmente, reducen su valor, degradando la característica importante.

# Ecuaciones (Explicación y Tablas):

El potencial electrico debido a una polarización es:
\begin{equation}
  \phi(r)=1/(4\pi \epsilon_0)*\int_T P(r')*(r-r')/|r-r'|^3dr'
\end{equation}
Momento dipolar:
\begin{equation}
P=1/\delta T \sum_{Pi \in \delta T} Pi
\end{equation}
Para las densidades de carga de polarización o carga ligada:
\begin{equation}
\rho_p = -\delta.P,\        \sigma=-n.[P]
\end{equation}
Para un par de dielectricos, si uno de ellos se encuentra en el vacio P=0, en terminos de $\rho \ y \ \sigma$ el potencial es:
\begin{equation}
\phi=1/(4\pi \epsilon_0) \int_T \rho_p/(|r-r'|)dr'+1/(4\pi \epsilon_0) \oint \sigma_p/(|r-r'|) dS'
\end{equation}
Empleando las ecuaciones de elestrostatica en presencia de dielectricos:
\begin{equation}
\nabla.E=(\rho_L+\rho_P)/\epsilon_0, \  \  \ \Delta \ x \ E=0
\end{equation}
Con salto:
\begin{equation}
n.[E]=(\sigma_L+\sigma_P)/\epsilon_0, \  \  \ n \ x \ [E]=0
\end{equation}
Siendo $P_L$ y $Q_L$ siendo las densidades de carga libre (no son de polarización) \
\
**Vector desplazamiento D**
Si se define 
$ D=\epsilon_0 E + P$
Entonces $\Delta.D=P_L$, y en salto, $n.[D]=\sigma_L$ \

**Dielectricos lineales y permitividad**
Para los dielectricos lineales tenemos una relación constitutiva P con E
\begin{equation}
P=\epsilon_0 X_e E,
\end{equation}
Siendo $X_e$ siendo la susceptibilidad del dielectrico.
Tambien se tiene que D y E son proporcinales:
\begin{equation}
D= \epsilon_0(1+X_e)E=\epsilon_0 \epsilon_r E = \epsilon E
\end{equation}
Donde $\epsilon_r$ es la permitividad relativa.
\
Donde $\epsilon$ es la permitividad absoluta.

Una particula electrica lineal adquiere un momento dipolar proporcional al campo aplicado y una fuerza que la empuja hacia donde el campo mas intenso.

P=\alpha E, \ \ \ F=(P.\nabla)E=\nabla(1/2 \ \alpha|E|^2)

**Energia electrica en presencia de dielectricos**
El valor de la energia puede calcularse si se conoce el camino que han seguido los dielectricos para polarizarse, si el desplazamiento va de 0 a D entonces tenemos energia en un punto y energia total.
\begin{equation}
u=\int_0^D EdD, \ \ \ \ U=\int u dt
\end{equation}
Integrando la parte anterior obtenemos 
\begin{equation}
u= 1/2 \epsilon E^2 = 1/2 E.D, \ \ \ \ U = \int udr
\end{equation}
donde la energia resultante incluye tanto el trabajo para reunir las cargas libres como el necesario para polarizar el dielectrico.
\
**Condensadores con dielectricos**
La capacidad de un condensador relleno con dielectrico:
\begin{equation}
C=Q_{L1}/(V_1-V_2)
\end{equation}
donde $Q_{L1}$ es la carga libre depositada en la superficie 1.
Para el caso de placas planas y paralelas con dielectrico homogeneo de permitividad $\epsilon$, la capacidad se obtiene:
\begin{equation}
C=\epsilon S/a
\end{equation}
La energia de un sistema de conductores inmersos en dielectricos lineales es tambien una extension de la correspondiente para el vacio.
\begin{equation}
U=1/2 \sum_i Q_{Li}V_i+1/2 \int \sigma_L \phi dS
\end{equation} 
**Campo electrico**
Cuando el campo electrico es muy alto, se arrancan electrones del dielectrico, este fenomeno se conoce como ruptura dielectrica.

Al maximo campo el material es capaz de soportar sin romperse se denomina rigidez dielectrica.

Para comprender mejor el tema tambien se necesita ecuaciones de campo electrico que ayuden a solucionar problemas conceptuales.

La magnitud del campo electrico producido por un capo de fuerza sobre una carga q:
\begin{equation}
E = F/q
\end{equation}
Siendo F la magnitud del campo de fuerza.
Para expresar la formula en terminos de la distancia:
\begin{equation}
E=Kq/d^2
\end{equation}
Donde $K=9*10^9 Nm^2/C^2$
\
Tambien se usaran valor de constante $\epsilon_0$ (constante de permitividad en el vacío)
\
Donde $\epsilon_0=8.85419 * 10^{-12} C^2/ Nm^2$

# Ejemplo e ilustración:

- Un condensador de placas paralelas tiene un área total de $6,45 * 10^{-4} m^2$ y una separación entre placas de $2 mm$ a través de las cuales se aplica un potencial de $12 V$. Si se interpone dentro de la región entre las placas un material con una constante dieléctrica $\epsilon_r$ de $6.0$,
añadido a esto una rigidez dieléctrica de $26 MV/m$. calcular:

**a)** La capacidad del condensador.

**b)** La magnitud de la carga almacenada en cada placa.

**c)** El desplazamiento dieléctrico D.

**d)** La polarización.

**e)** El potencial eléctrico que podrá soportar sin perforación.

**Sln//** Formulas a usar:

$C = \epsilon_0 \epsilon_r A/d$  (1)

$Q=C.V$ (2)

$D=\epsilon  E = \epsilon_0 \epsilon_r V/e$ (3)

$D= \epsilon_0 E+P$ (4)

$V= E_R d$ (5)

Donde $E_R$ es la rigidez dielectrica




```
%matplotlib inline
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
e0 = 8.85419e-12      #comenzamos definiendo las variables que conocemos
er = 6.0
Er = 26e6
A = 6.45e-4
d = 2e-3
V = 12
print("Respuestas:")
C = e0*er*A/d         #para calcular la parte "a" usamos la formula vista antes y calculamos la capacidad 
print( "(a)","C =",C,"F")
Q = C*V               #ya que hemos calculado la capacidad podemos calcular la carga
print("(b)","Q =",Q,"C")      
D = e0*er*V/d         #el desplazamiento dielectrico se calculara mediante esta ecuación
print("(c)","D =",D,"C/m^2")
#despejando la ecuacion en terminos de la polarización (4) obtenemos,
P = D-e0*V/d
print("(d)","P =",P,"C/m^2")
#continuando con el punto e, el potencial que soportara se define por,
Vs = Er*d
print("(e)","Vs =",Vs,"v")
#Para este problema tenemos un campo electrico de,
Em = D/(e0*er)
print("Em=",Em,"N/C",", campo electrico que se interpone entre las placas el material dielectrico.")
#Asumiendo que antes de que se interponiese el dielectrico hubiese habido aire tendriamos un campo:
Eaire = (e0*(1.00054)*V/d)/(e0*(1.00054))
print("Eaire=",Eaire,"N/C", ", campo electrico si el dielectrico el medio fuese aire")
print("Como Em es igual a Eaire se asume que el campo electrico es invariante segun el dielectrico que se use.")
```
